```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
# import data
data1 = pd.read_csv('data/task1.csv')
```

```python
# Showing top 5 countries
data1 = data1.head(5)

data1.plot(kind='barh', x='currency_name', y='amount_gbp_millions', color='#3366FF')
plt.xlabel('')
plt.ylabel('')
plt.title('Total card spend (millions of GBP, July 2023)', y=1.02, fontsize=15)

plt.gca().invert_yaxis()

# add numbers on the bars
for index, value in enumerate(data1.head(5)['amount_gbp_millions']):
    plt.text(value, index, str(value)+'m')
# xlim to fit labels on bars
plt.xlim(0, 67)

# Add line at 5 
plt.axvline(x=5, color='red', linestyle='--')

plt.legend(['5m threshold', 'Total card spend'])

```
