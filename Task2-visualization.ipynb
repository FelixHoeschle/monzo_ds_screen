```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
# import data
data2 = pd.read_csv('data/task2.csv')
```

```python
data2.set_index(pd.to_datetime(data2['date']), inplace=True)
data2.drop('date', axis=1, inplace=True)
```

```python
currency_dict = {
    '356': 'Indian Rupee',
    '986': 'Brazilian Real',
    '392': 'Japanese Yen'
}
```

```python
# line plot for mid_market_rate_7d_avg for each currency_name
fix, axes = plt.subplots(3, 1, figsize=(12, 5))
for i, cur in enumerate(currency_dict.keys()):
    axes[i].plot(data2.loc[data2['ISO4217_currency_code']==int(cur)]['mid_market_rate_7d_avg'], color='#3366FF')
    axes[i].set_title(currency_dict[str(cur)])

plt.tight_layout()
```
